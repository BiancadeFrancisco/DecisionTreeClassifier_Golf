```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
golf = pd.read_excel("/content/golf_df.xlsx")
golf.head()
```

```python
x = pd.get_dummies(golf[["clima", "temp", "umidade", "vento"]])

x.head()
```

```python
y = golf[["jogar"]]

y.head()

# essa variável podemos utilizar ela categórica, uma das vantagens da arvore de decisão
```

```python
# como a base de dados é pequena, não vamos dividir em treino e teste!
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
```

```python
modelo = arvore.fit(x, y)
modelo
```

```python
fig = plt.figure(figsize= (70, 30))

fig = tree.plot_tree(modelo,
                     feature_names  = x.columns,
                     class_names = ["yes", "no"],
                     filled = True)
```
